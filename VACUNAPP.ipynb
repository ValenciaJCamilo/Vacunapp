```python
# ...............................................................................................................................................................................
# PROYECTO INTEGRADOR - VACUNAPP
# ESTUDIANTES: VALENCIA SILVA JUAN CAMILO, PARADA SEVERICHE GABRIEL FELIPE, 
#VILLAMIZAR M√ÅRQUEZ CLAUDIA TATIANA
# ...............................................................................................................................................................................

# ...............................................................................................................................................................................
# Inicialmente, se tienen nodos de origen y nodos de destino. Estos nodos en el
# mapa, representar√°n los lugares 
#de entrega a donde se repartir√°n las vacunas del COVID-19
# ...............................................................................................................................................................................

nodoorigen = ('A','B','C','D','E','F','G','H','I','J','K','L','M','N')
nododestino = ('A','B','C','D','E','F','G','H','I','J','K','L','M','N')

# ...............................................................................................................................................................................
# En este apartado es donde se crea el grafo. Primeramente se define el grafo, 
#posteriormente se procede a hacer la conexi√≥n con sus nodos vecinos, a la vez 
#que se establecen los 
# pesos, que en este caso ser√≠a la distancia entre cada esquina. 
# En el nodo A, por ejemplo, tiene relaci√≥n con el nodo D, que tiene un costo de
# 174 metros y una relaci√≥n con el nodo E, que tiene costo de 145 metros.
# ...............................................................................................................................................................................

grafo = {
'A':{'D':174,'E':145},
'B':{'E':337, 'H':174},
'C':{'J':244,'N':364},
'D':{'E':168,'B':365,},
'E':{'A':145,'D':168,'G':327},
'F':{'E':308,'C':244},
'G':{'E':327,'M':105,'I':450,'L':236},
'H':{'D':333,},
'I':{'B':148,'L':309},
'J':{'F':197,'C':244},
'K':{'L':210},
'L':{'K':210,'G':236,'M':255},
'M':{'N':134,'G':105,'K':186},
'N':{'M':134,'C':364}
}

# ...............................................................................................................................................................................
# Aqu√≠ se define el algoritmo a usar, que en este caso ser√° Dijkstra. 
# ...............................................................................................................................................................................

def dijkstra(graph,start,goal):
# Necesario para tener registro del coste de llegada a un nodo. Este ser√° muy
# importante debido a que se actualizar√° varias veces en el algoritmo.
    distancia_minima = {} 
# Necesario para llevar el registro del camino que llev√≥ a llegar al nodo en 
#cuesti√≥n.
    rastreo_nodo_predecesor = {} 
# Necesario para iterar varias veces a trav√©s de todos los nodos del grafo.
    iterNodos = graph 
# Se define infinito, que puede llegar a considerarse como un n√∫mero bastante
# grande en el grafo.
    infinito = 9000 
# Necesaria, ya que ser√° la encargada de guardad y mostrar el camino de la ruta 
#m√°s √≥ptima.
    ruta_final = [] 

# ...............................................................................................................................................................................
# Primeramente, se asigna a 0 como costo de inicio para llegar al nodo origen 
#, y se establece infinito (en nuestro caso 9000)
# como el costo para alcanzar a todos los dem√°s nodos.
# Este bucle seguir√° corriendo hasta que todos los nodos se hayan visitado.
# ...............................................................................................................................................................................

    for node in iterNodos:
        distancia_minima[node] = infinito

    distancia_minima[start] = 0

# ...............................................................................................................................................................................
# Para poder iterar en el grafo, es necesario determinar el nodo con la menor 
#distancia cada vez que se avanza. Esto es muy importante, ya que de no hacerlo
# el algoritmo no tendr√≠a sentido alguno.
# ...............................................................................................................................................................................

    while iterNodos:
        nodo_minima_distancia = None

        for node in iterNodos:
            if nodo_minima_distancia is None:
                nodo_minima_distancia = node

            elif distancia_minima[node] < distancia_minima[nodo_minima_distancia]:
                nodo_minima_distancia = node

# ...............................................................................................................................................................................
# Una vez teniendo el nodo con un costo m√≠nimo, se procede a determinar los posibles 
#caminos, es decir, los nodos vecinos del nodo actual.
# ...............................................................................................................................................................................
       
        path_options = graph[nodo_minima_distancia].items()

# ...............................................................................................................................................................................
# Para seguir con el algoritmo, se debe calcular el costo cada vez que se avanza
# en el grafo, de forma que se actualice √∫nicamente
# si el costo existente es menor
# ...............................................................................................................................................................................

        for nodo_predecesor, peso in path_options:
            if peso + distancia_minima[nodo_minima_distancia] < distancia_minima[nodo_predecesor]:
                distancia_minima[nodo_predecesor] = peso + distancia_minima[nodo_minima_distancia]
                rastreo_nodo_predecesor[nodo_predecesor] = nodo_minima_distancia

# ...............................................................................................................................................................................
# En esta secci√≥n de c√≥digo, se procede a retirar los nodos que se acaban de visitar para no volver a 
#iterar sobre estos, ya que no ser√° necesario pasar dos veces.
# El m√©todo pop, nos ayuda a eliminar y retornar un elemento en la lista, que en este caso ser√≠a el nodo 
#con la m√≠nima distancia, ya que es el que nos interesa.
#............................................................................................................................................................................
        iterNodos.pop(nodo_minima_distancia)
# ...............................................................................................................................................................................
# Ahora que se ha llegado finalmente al nodo deseado, se procede a calcular el costo total acumulado, es 
#decir, se calculan los metros totales recorridos.
# ...............................................................................................................................................................................

    currentNode = goal

    while currentNode != start:

        try:
            ruta_final.insert(0,currentNode)
            currentNode = rastreo_nodo_predecesor[currentNode]
        except KeyError:
            print('No es posible trazar una ruta. Por favor ejecute nuevamente e ingrese otro nodo destino.')
            break
    ruta_final.insert(0,start)
  
    if distancia_minima[goal] != infinito:

# ...............................................................................................................................................................................
# Se procede a imprimir la ruta y la distancia total. 
# ...............................................................................................................................................................................

        print('La ruta a seguir para la distribuci√≥n de las vacunas es la siguiente üó∫Ô∏è: ' + str(ruta_final))
        print('La distancia total ser√° de üöö: ' + str(distancia_minima[goal])+ ' metros')


# ...............................................................................................................................................................................
# Se le pide al usuario que ingrese el nodo destino y origen. 
# ...............................................................................................................................................................................


print('Bienvenido al sistema de ruteo √≥ptimo para la manipulaci√≥n y ') 
print ('distribuci√≥n de las vacunas COVID-19 en Colombia üíâ.')
print('¬øDe qu√© ubicaci√≥n salen las vacunas?')
nodoorigen = input()
print('¬øHacia d√≥nde se distribuir√°n las vacunas?')
nododestino = input()

# ...............................................................................................................................................................................
# Finalmente, se establece que el nodo origen y nodo destino ser√° el que digite el usuario. 
# ...............................................................................................................................................................................

dijkstra(grafo,nodoorigen,nododestino)
    
```
